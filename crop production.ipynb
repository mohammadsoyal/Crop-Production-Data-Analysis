```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\moham\Downloads\Crop Production data.csv")
df
```

```python
# Fill missing values in the 'Production' column with the median value
df['Production'].fillna(df['Production'].median(), inplace=True)
```

```python
# Remove duplicate rows
df_cleaned = df.drop_duplicates()
```

```python
# Verify the cleaned df
print(df_cleaned.info())
print(df_cleaned.head())
```

```python
df_cleaned.to_csv('cleaned crop production data.csv')
df_cleaned
```
